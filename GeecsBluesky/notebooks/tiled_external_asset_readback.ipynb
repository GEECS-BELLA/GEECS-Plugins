# Tiled External Asset Readback

Use Tiled as the archive source of truth for a Bluesky scan, then fill a native GEECS camera asset locally. The selected shot is matched by `scan_event_index`; the image filename is resolved from the event's device `acq_timestamp`.

```python
from geecs_bluesky.assets.tiled_readback import load_camera_image_from_tiled
```

## Parameters

Set these values to a Bluesky-created scan recorded in Tiled. Leave `DEVICE_TYPE = None` to query the live GEECS database. `TILED_URI` and `TILED_API_KEY` default to the shared `~/.config/geecs_python_api/config.ini` values.

```python
RUN_TILED_CAMERA_LOOKUP = False

EXPERIMENT = "Undulator"
YEAR = 2026
MONTH = 6
DAY = 23
SCAN_NUMBER = 1
DEVICE_NAME = "UC_Amp2_IR_input"
SHOT_NUMBER = 1

DEVICE_TYPE = None
TILED_URI = None
TILED_API_KEY = None
ROOT_MAP = None
```

```python
loaded = None
camera_image = None

if RUN_TILED_CAMERA_LOOKUP:
    loaded = load_camera_image_from_tiled(
        year=YEAR,
        month=MONTH,
        day=DAY,
        scan_number=SCAN_NUMBER,
        device_name=DEVICE_NAME,
        shot_number=SHOT_NUMBER,
        experiment=EXPERIMENT,
        device_type=DEVICE_TYPE,
        tiled_uri=TILED_URI,
        tiled_api_key=TILED_API_KEY,
        root_map=ROOT_MAP,
        retry_intervals=[],
    )
    camera_image = loaded.image
    start = loaded.asset.start_doc
    print(f"run uid: {start.get('uid')}")
    print(f"scan folder: {start.get('scan_folder')}")
    print(f"data key: {loaded.asset.data_key}")
    print(f"datum id: {loaded.asset.datum_id}")
    print(f"acq_timestamp: {loaded.asset.acq_timestamp:.3f}")
    print(f"file path: {loaded.asset.file_path}")
    print(f"image: {camera_image.shape} {camera_image.dtype}")
else:
    print("Set RUN_TILED_CAMERA_LOOKUP = True after entering a real Bluesky scan.")
```

```python
if camera_image is not None:
    import matplotlib.pyplot as plt

    plt.figure(figsize=(6, 6))
    plt.imshow(camera_image, cmap="gray")
    plt.axis("off")
```
